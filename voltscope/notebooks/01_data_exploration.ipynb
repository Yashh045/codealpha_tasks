## 1. Load & Inspect Raw Data

```python
df = load_processed_data()

print(f"Shape: {df.shape}")
print(f"Countries: {df['country'].nunique()} | Years: {df['year'].min()}–{df['year'].max()}")
df.head(10)
```

## 2. Summary Statistics

```python
df.describe().round(2)
```

## 3. Quick Visual — Renewable Share by Country (2024)

```python
latest = df[df["year"] == 2024].sort_values("renewable_share_pct", ascending=True)

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(latest["country"], latest["renewable_share_pct"], color=COLORS["accent"])
ax.set_title("Renewable Electricity Share — 2024")
ax.set_xlabel("Renewable Share (%)")
ax.axvline(50, color=COLORS["highlight"], linestyle="--", alpha=0.6, label="50% target")
ax.legend()
plt.tight_layout()
plt.show()
```
